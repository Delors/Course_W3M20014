$X_{n+1} = (aX_n + c)\; mod\; m$

```python
def lcg(x0, a,c,m, count):
    print(f"seed={x0}, a={a}, c={c}, m={m}")
    i = 0
    xn = x0
    while i < count:
        print(f"{xn}, ", end="")
        xn = (a * xn + c ) % m
        
        i += 1 
```

```python
lcg(1234,-11,7,113,100)

lcg(1234,8,8,256,100)

lcg(1234,-8,8,256,100)
```
